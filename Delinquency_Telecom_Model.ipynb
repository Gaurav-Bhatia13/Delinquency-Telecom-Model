```python
#Considering values as outliers whose z-score > threshold 
```

```python
from scipy.stats import zscore
threshold = 3
def outL(data):
  db['zscore-cal'] = np.abs(zscore(data))
  threshold = 3
  outliers = db[db['zscore-cal'] > threshold] or db[db['zscore-cal'] < threshold]
  return len(outliers)
```

```python
db.describe()
```

```python
#Removing entries that are containing outliers
```

```python
for col in db.columns:
    # Z-scores for the column
    z_scores = zscore(db[col])
    # identify values with Z-scores greater than or less than 3
    outliers = (z_scores > 3) | (z_scores < -3)
    # drop rows with outlier values
    db = db[~outliers]
```

```python
from imblearn.over_sampling import SMOTE
```

```python
xd = dd.drop(labels=["label"],axis=1)
yd = dd["label"]
```

```python
#Initialise smote - Synthetic Minority Over-sampling Technique
smote = SMOTE()

x_smote, y_smote = smote.fit_resample(xd,yd)

print('old shape', xd.shape)
print('new shape', x_smote.shape)
```

```python
x_smote.describe()
```

```python
y_smote = pd.DataFrame(y_smote, columns =['label'])
y_smote.head()
```

```python
#Visualise data imbalance
import seaborn as sns
plt.figure(figsize=(5,5))
sns.countplot('label',data=y_smote,palette='hls')
plt.show()
```

```python
#applying logistic regression
from sklearn.model_selection import train_test_split
xd_train, xd_test, yd_train, yd_test = train_test_split(x_smote, y_smote, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
modelLRd = LogisticRegression()
modelLRd.fit(xd_train, yd_train)
print('Training score =', modelLRd.score(xd_train, yd_train))
print('Test score =', modelLRd.score(xd_test, yd_test))

from sklearn.metrics import confusion_matrix
ypredd = modelLRd.predict(xd_test)
cmd = confusion_matrix(yd_test, ypredd)
tn = cmd[0,0]  #True Negative
tp = cmd[1,1]  #True Positives
fp = cmd[0,1]  #False Positives
fn = cmd[1,0]  #False Negatives

accuracy = (tp+tn)/(tp+fn+fp+tn)
precision = tp / (tp+fp)
recall = tp / (tp+fn)
f1 = 2*precision*recall / (precision+recall)

print('Accuracy =',accuracy)
print('Precision =', precision)
print('Recall =', recall)
print('F1 Score =', f1)
```
